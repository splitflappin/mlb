```python
import pandas as pd
import pybaseball as pyb
```

```python
def loadSav(start_date, end_date):
  # format = YYYY-MM-DD
  sav = pyb.statcast(start_dt=start_date, end_dt=end_date)
  return sav
```

```python
sav25 = loadSav("2025-01-01", "2025-12-31")
sav25.head()
```

```python
sav25.to_csv("../data/season2025.csv", index=False)
```
